# Motif Manual Retrieval Smoke Test

Use this notebook after `docker compose up` and ingestion. It verifies that a theme query returns chunks with film and source metadata.

```python
import weaviate

client = weaviate.connect_to_local(host='localhost', port=8080)
collection = client.collections.get('MotifChunk')
result = collection.query.near_text(query='doubling identity performance', limit=8)
for obj in result.objects:
    print(obj.properties['film_slug'], obj.properties['source_key'])
    print(obj.properties['text'][:400], '\n')
client.close()
```
